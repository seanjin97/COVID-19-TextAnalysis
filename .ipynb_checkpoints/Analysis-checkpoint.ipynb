```python
def correct_spelling(text): #string, list of words to not change
    temp = []
    for word in text.split(' '):
        if (0.985 <= Word(word).spellcheck()[0][1] < 1.0):
            w = Word(word).spellcheck()[0][0]
            temp.append(w)
        else:
            temp.append(word)
            
    return " ".join(temp)
```

### Faster performance text preprocessing ###

```python
def clean_text(text):
    text = text.translate(tab)
    # tokenize
    text = text.split(" ")
    # remove words that contain numbers
#     text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    text = [x for x in text if x not in stop_list]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos[t[1][0]]) for t in pos_tags]
    # remove stop words    
    text = [x for x in text if x not in stop_list]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
#     text = " ".join(text)
    return(text)
```

# Model tuning

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
#         model=LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
        model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=dictionary,
                                       num_topics=num_topics, 
                                       random_state=0,
                                       chunksize=100,
                                       passes=100,
                                       per_word_topics=True)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus_vect, texts=tokenised_words, start=2, limit=40, step=6)
# Show graph
import matplotlib.pyplot as plt
limit=40; start=2; step=6;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

# Overall results

We observe that there are some overlaps of what netizens are talking about on the forums for both of the LDA models generated. From the general sentiment analysis done, we found that a large majority of comments were neutral with a minority having negative sentiments. This matches up with the topics modelled as a majority of them refer to the reports of number of cases/ deaths and government meaures reported which are generally neutral in nature since they are just discussions on the news reports. As the COVID-19 situation is one of the current hottest topics in Singapore, it is no surprise that these news related posts garner a large amount of online attention which results in more people talking about them. 

However, we are still able to identify topics that have negative sentiments such as the first few imported cases from Chinese tourists that garnered a large amount of negative criticism online. Concerns of the unavailability of masks being sold in retail shops also sparked concerns amongst citizens as they worry of being insufficiently prepared for the virus situation. 

Overall, this analysis serves as a simple prelimnary analysis to find out what netizens are talking about on the topic of the COVID-19 situation. In order to extract out deeper insights such as the opinions on government measures, more time needs to be put into fine tuning the models such as the identification of more stopwords to better filter out more commonly used words that do not add value and better model hyperparameter tuning.
